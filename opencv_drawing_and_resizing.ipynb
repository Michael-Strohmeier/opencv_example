```python
import cv2

# using this method instead of "cv2.imshow" -> cv2.imshow(img) -> because it crashes google colab
from google.colab.patches import cv2_imshow
```

```python
path = "/content/Ek_Oky5U8AAmoCC.jpeg"  
img = cv2.imread(path)
```

```python
# scale image

scale_percent = 30
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)

resized_img = cv2.resize(img, dim, interpolation=cv2.INTER_AREA)
```

```python
# cv2.imshow(resized_img)
cv2_imshow(resized_img)
```

```python
# drawing rectangle

start_point = (15, 30)
end_point = (50, 65)

# Blue, Green, Red
color = (255, 0, 0)
thickness = 2

temp_img = cv2.rectangle(resized_img, start_point, end_point, color, thickness)
cv2_imshow(temp_img)
```
